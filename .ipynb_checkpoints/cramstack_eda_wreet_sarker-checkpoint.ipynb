```python
# Let's import the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# First let's check the game sales data by converting it into a dataframe

game_sales_df = pd.read_csv("game_sales_data.csv")
game_sales_df.info()
```

```python
# Checking the first 5 values of the dataset

game_sales_df.head()
```

```python
# Checking the bottom 5 values of dataset

game_sales_df.tail()
```

```python
# Checking how many unique Platforms are in the data set and their count

game_sales_df["Platform"].value_counts()
```

```python
# Creating a new modified dataframe that contains only "PC", "PS4", "PS3", "XOne", "X360"

game_sales_modified = game_sales_df[game_sales_df["Platform"].isin(["PC", "PS4", "PS3", "XOne", "X360"])]
```

```python
# Checking the modified dataframe

game_sales_modified
```

```python
game_sales_modified.info()
```

```python
game_sales_modified["Platform"].value_counts()
```

# 3. Visualization of average critic score per developer

```python
game_sales_modified
```

```python
# Grouping the data by the developers

game_sales_modified.groupby("Developer").mean()["Critic_Score"]
```

```python
grouped_by_developer = game_sales_modified.groupby("Developer").mean()
```

```python
grouped_by_developer.reset_index()["Developer"]
```

```python
developers = grouped_by_developer.reset_index()["Developer"]
critic_vals = game_sales_modified.groupby("Developer").mean()["Critic_Score"]
```

```python
# Scatter plot to show Average critic score per developer

plt.figure(figsize=(20, 20))
plt.scatter(developers,critic_vals,
            alpha=0.4, s = critic_vals)
```

# 4. Top 5 games with biggest differences in critic vs user score

```python
# Creating a new column that contains the difference between critic score and user score

game_sales_modified["Critic_User_Difference"] = game_sales_modified["Critic_Score"]- game_sales_modified["User_Score"]
```

```python
game_sales_modified
```

```python
# Sorting the dataframe by the critic and user score difference column and showing the top 5 values

game_sales_modified.sort_values(by="Critic_User_Difference", ascending=False).head()
```
